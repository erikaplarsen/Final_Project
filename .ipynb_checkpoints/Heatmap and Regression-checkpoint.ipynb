```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
LoanData = pd.read_csv("Loan Data 2.csv")
```

```python
LD = LoanData[LoanData!= '..']
LD1 = LD.dropna()
LD1.head()
LD1.info()
LD1.describe()
LD1.columns
```

```python
sns.distplot(LD1['Loan Amount'])
```

```python
LD2 = LD1.corr()
```

```python
sns.heatmap(LD2)
```

```python
X = LD1[['Funded Amouint', 'Installment', 'Annual Income',
               'DTI', 'Delinquencies last 2-yrs','# Inquiries last 6 months', 'Open Accounts', 'Revolving Balance',
         'Total # of Accounts', 'Total Debt Owed']]
y = LD1['Loan Amount']
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1000)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)
print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
predictions = model.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```
